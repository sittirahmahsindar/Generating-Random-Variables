```python
import numpy as np
import matplotlib.pyplot as plt
```

**Parameter distribusi binomial**

```python
n = 1000       # jumlah produksi per hari
p = 0.02       # probabilitas cacat
days = 30      # jumlah hari simulasi
```

**Simulasi binomial**

```python
# Simulasi jumlah barang cacat dalam 30 hari produksi
responses = np.random.binomial(n, p, 30)
print(responses)
```

**Histogram**

```python
plt.figure(figsize=(8, 5))
plt.hist(responses, bins=range(0, 50), density=True, alpha=0.7, color="b", edgecolor="black", align="left")
plt.xlabel("Jumlah Barang Cacat per Hari")
plt.ylabel("Frekuensi Relatif")
plt.title("Simulasi Jumlah Barang Cacat (Distribusi Binomial)")
plt.xticks(range(0, 50, 5))
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

Histogram menunjukkan bahwa distribusi jumlah barang cacat per hari masih membentuk pola yang mendekati distribusi normal, dengan puncak utama berada di sekitar nilai 20. Hal ini menandakan bahwa hasil simulasi tetap konsisten dengan nilai rata-rata yang mencerminkan distribusi binomial. Sebagian besar data disajikan pada kisaran 16 hingga 25 barang cacat per hari. Hal ini menunjukkan bahwa nilai cacat harian tidak menyebar terlalu jauh, namun tetap memiliki variasi di sekitar rata-rata. Meskipun demikian, terdapat perbedaan frekuensi tinggi antar nilai yang menunjukkan adanya variasi yang cukup terlihat. Beberapa nilai muncul lebih sering dibandingkan yang lain, sehingga distribusi tidak sepenuhnya merata.

Secara keseluruhan, jumlah barang cacat per hari menunjukkan variasi yang cukup besar sehingga tidak sepenuhnya stabil, namun masih berada dalam batas yang wajar sesuai dengan karakteristik distribusi binomial.

**Rata-rata & Standar deviasi**

```python
mean_cacat = np.mean(responses)
std_cacat = np.std(responses)

print("\nRata-rata barang cacat per hari:", mean_cacat)
print("Standar deviasi:", std_cacat)
```

Rata-rata jumlah barang cacat per hari adalah sekitar 19.67, yang sangat mendekati dengan nilai teoritis 20, sehingga hasil simulasi sesuai dengan distribusi binomial. Standar deviasi sebesar 3.17 menunjukkan bahwa penyebaran data tidak terlalu besar dari rata-rata.

Jadi sebagian besar jumlah cacat berada di kisaran sekitar 16–23 per hari, sehingga variasinya relatif kecil. Oleh karena itu, dapat disimpulkan bahwa jumlah barang cacat per hari cenderung stabil dengan kerusakan yang kecil .

**Hasil analisis**

```python
print("\nAnalisis:")
if std_cacat < mean_cacat * 0.2:
    print("Jumlah barang cacat per hari cenderung konsisten dan tidak mengalami perubahan yang signifikan dari rata-ratanya.")
else:
    print("Jumlah barang cacat per hari menunjukkan variasi yang cukup besar sehingga tidak bersifat stabil.")
```
