#**Screping web dari halaman 1–5**

```python
from bs4 import BeautifulSoup
from requests import get
```

```python
url = "https://loginmegastore.com/search?query=kulkas&p=true&page=3"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

Title Barang

```python
all_title_barang = []

for page in range(1, 6):
    url = f"https://loginmegastore.com/search?query=kulkas&p=true&page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    title_barang = soup.find_all("p", class_="mantine-focus-auto !leading-[120%] m_b6d8b162 mantine-Text-root")

    print(f"=== HALAMAN {page} (Title Barang) ===")

    for x in title_barang:
        print(x.prettify())

    for cek_title_barang in title_barang:
        print(cek_title_barang.text)
        all_title_barang.append(cek_title_barang.text)
```

Harga Barang

```python
all_harga_barang = []

for page in range(1, 6):
    url = f"https://loginmegastore.com/search?query=kulkas&p=true&page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    harga_barang = soup.find_all("p", class_="mantine-focus-auto !whitespace-nowrap m_b6d8b162 mantine-Text-root")

    print(f"=== HALAMAN {page} (Harga Barang) ===")

    for x in harga_barang:
        print(x.prettify())

    for cek_title_barang in harga_barang:
        print(cek_title_barang.text)
        all_harga_barang.append(cek_title_barang.text)
```

Gambar Barang

```python
all_gambar_barang = []

for page in range(1, 6):
    url = f"https://loginmegastore.com/search?query=kulkas&p=true&page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    gambar_barang = soup.find_all("img", class_="m_9e117634 mantine-Image-root")

    print(f"=== HALAMAN {page} (Gambar Barang) ===")

    for x in gambar_barang:
        print(x.prettify())

    for cek_title_barang in gambar_barang:
        print(cek_title_barang)
        all_gambar_barang.append(cek_title_barang)
```

#**Simpan ke format json**

```python
import json
min_len = min(len(all_title_barang), len(all_harga_barang), len(all_gambar_barang))

List_Elektronik = []
for i in range(min_len):
    item = {
        "Nama Barang": all_title_barang[i].strip(),
        "Harga": all_harga_barang[i].strip(),
        "Gambar": all_gambar_barang[i].get('src') if all_gambar_barang[i] else None
    }
    List_Elektronik.append(item)

with open("List_Elektronik.json", "w", encoding="utf-8") as f:
    json.dump(List_Elektronik, f, ensure_ascii=False, indent=4)

print("File JSON berhasil dibuat")
```

# **Prinsip ETL (format JSON ke CSV)**

Transformasi

```python
import pandas as pd

df = pd.DataFrame(List_Elektronik)

# Bersihkan harga (hapus Rp dan titik)
df["Harga"] = df["Harga"].str.replace("Rp", "", regex=False)
df["Harga"] = df["Harga"].str.replace(".", "", regex=False)

# Ubah ke numerik
df["Harga"] = pd.to_numeric(df["Harga"], errors="coerce")

# Hapus data kosong
df = df.dropna()

df.head()
```

Simpan ke CSV

```python
df.to_csv("List_Elektronik_clean.csv", index=False)
print("File CSV berhasil dibuat")
```
